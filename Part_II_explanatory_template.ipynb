```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('newdf.csv')
```

```python
dff1 = df.dropna(subset=['Age']).copy()
dff1['Age']=dff1['Age'].astype(int)
```

## Visualization 1


This visualization uses a countplot that explicitly divides our customers/subscribers amongst the 3 recognized Gender roles


```python
plt.figure(figsize=(15,6))
sns.countplot(data=df, x='member_gender', hue='user_type');
plt.xlabel('Genders')
plt.ylabel('Count')
plt.title('UserType-Gender Counts');


#we aim to use a countplot with hue representing our unique usertypes for distinguish between all 3 genders
```

* In the barplot above, we see the distribution of Genders in our entire dataset.
* Safe to say it is in a descreasing order; from Males, Females and then Others

## Visualization 2

Interestingly, we try and see how our different Ages having different riding timeframes are either Customers or Subscribers

Observations:
* As our Age increases, we tend to see that the Duration things out, visualizing a negative correlation there
* So also do we see that both user_types are spread across all ages really


```python
#using a dodge paramater, very small jitter size and marker size of 2, 
#we can distinctly see how our unique usertype-ages are represented 

plt.figure(figsize=(15,6))
sns.stripplot(data=dff1, y='duration_sec', x='Age', hue='user_type', dodge=True, jitter=0.35, size=2)
plt.title('Different usertypes, their Ages and cycling Duration(in s)')
plt.ylabel('Duration in seconds (s)')
plt.xlabel('Age')
plt.xticks(rotation=90);
#90 rotation to make viewing much easier
```

* From these dataset we can see how the duration in seconds of ALL user ages thin out as the ages go higher. So also do we have a very sparse distribution when we start getting to about 80000 seconds-mark.

* Although our data is properly dominated by subscribers, it is fair to say occasional customers too hold their own up when it comes to cycling really long distances

## Visualization 3

By far our most important visualization, of course Age, Gender and others variables count but this here, bike_share_for_all just goes to show if or not our users are occasional customers or actual subscribers.



```python
plt.figure(figsize=(15,6))
sns.countplot(data=df, hue='user_type', x='bike_share_for_all_trip')
plt.xlabel('Enrolled in Bike Share Program')
plt.ylabel('Count')
plt.title('Bike Share Program vs UserType');


#this is made with a hue of our targer variable, 'usertype'. aiming to visualize its distribution in our bike share program
```

Above we can see the difference in distribution using a Countplot(for counting categorical data) between people enrolled and those not enrrolled in our bike shraing program.
